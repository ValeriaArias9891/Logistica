```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df
```

```python
df[['Age','Survived']].head()
```

```python
df[['Survived','Age']].plot.scatter(x='Age',y='Survived')
```

```python
w= 0.09
b= -3.6


x= np.linspace(0, df['Age'].max(), 100)
y= 1/(1+np.exp(-(w*x+b)))
```

```python
df.plot.scatter(x='Age',y='Survived')
plt.plot(x,y, color='black')
plt.ylim(0,df['Survived'].max()*1.1)
plt.scatter(x,y, color='pink')

plt.xlabel('Age')
plt.ylabel('Supervivencia 1: positivo, 0: negativo')
plt.show
```
